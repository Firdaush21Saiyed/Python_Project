**5. Games with Contrasting Popularity by Region**

Na sales = North America,

Eu sales = Europ Sales,

Jp sales = Japan Sales,

Global sales = Total Worldwide sales

```python
pd.options.display.float_format = '{:.1f}'.format
```

```python
vg["NA_ratio"]=vg["NA_Sales"]/vg["Global_Sales"]
```

```python
vg["EU_ratio"]=vg["EU_Sales"]/vg["Global_Sales"]
```

```python
vg["JP_ratio"]=vg["JP_Sales"]/vg["Global_Sales"]
```

```python
vg["Other_ratio"]=vg["Other_Sales"]/vg["Global_Sales"]
```

```python
vg.head()
```

**5.1  North AmericaDominant Games with Weak Sales in Other Regions**





```python
na_popular = vg[(vg.NA_ratio>0.8) & (vg.EU_ratio<0.2) & (vg.JP_ratio<0.2)& (vg.Other_ratio<0.2)]
```

```python
na_popular=na_popular.sort_values(by="NA_ratio",ascending=False)
na_popular
```

```python
data = na_popular.head(5)
names = data['Name']
na = data['NA_ratio']
eu = data['EU_ratio']
jp = data['JP_ratio']
other = data['Other_ratio']

plt.figure(figsize=(7,5))

plt.bar(names, na,width=0.5, label="NA")
plt.bar(names, eu,width=0.5, bottom=na, label="EU")
plt.bar(names, jp,width=0.5, bottom=na+eu, label="JP",color="yellow")
plt.bar(names, other,width=0.5, bottom=na+eu+jp, label="Other")

plt.xlabel("Game Name")
plt.ylabel("Sales Ratio")
plt.title("Sales Performance of NAs Top 5 Games Across Global Regions")
plt.xticks(rotation=90)
plt.legend()

plt.show()
```

**5.2 EuropeDominant Games with Weak Sales in Other Regions**

```python
eu_popular = vg[(vg.EU_ratio>0.8) & (vg.NA_ratio<0.2) & (vg.JP_ratio<0.2)& (vg.Other_ratio<0.2)]
eu_popular=eu_popular.sort_values(by="EU_ratio",ascending=False)
eu_popular
```

```python
data = eu_popular.head(5)
names = data['Name']
na = data['NA_ratio']
eu = data['EU_ratio']
jp = data['JP_ratio']
other = data['Other_ratio']

plt.figure(figsize=(7,5))

plt.bar(names, eu,width=0.5, label="EU")
plt.bar(names, na,width=0.5, bottom=eu, label="NA")
plt.bar(names, jp,width=0.5, bottom=eu+na, label="JP")
plt.bar(names, other,width=0.5, bottom=eu+na+jp, label="Other")

plt.xlabel("Game Name")
plt.ylabel("Sales Ratio")
plt.title("Sales Performance of Europes Top 5 Games Across Global Regions")
plt.xticks(rotation=90)
plt.legend()

plt.show()
```

**5.3 JapanDominant Games with Weak Sales in Other Regions**

```python
jp_popular = vg[(vg.JP_ratio>0.8) & (vg.NA_ratio<0.2) & (vg.EU_ratio<0.2)& (vg.Other_ratio<0.2)]
jp_popular=jp_popular.sort_values(by="JP_ratio",ascending=False)
jp_popular
```

```python
data = jp_popular.head(5)
names = data['Name']
na = data['NA_ratio']
eu = data['EU_ratio']
jp = data['JP_ratio']
other = data['Other_ratio']

plt.figure(figsize=(7,5))

plt.bar(names, jp,width=0.5, label="JP",color="green")
plt.bar(names, na,width=0.5, bottom=jp, label="NA")
plt.bar(names, eu,width=0.5, bottom=jp+na, label="EU")
plt.bar(names, other,width=0.5, bottom=jp+eu+na, label="Other",color="orange")

plt.xlabel("Game Name")
plt.ylabel("Sales Ratio")
plt.title("Sales Performance of Japans Top 5 Games Across Global Regions")
plt.xticks(rotation=90)
plt.legend()

plt.show()
```

**5.4 Other RegionDominant Games with Weak Sales in NA, Europe, and Japan**

```python
Other_popular = vg[(vg.Other_ratio>0.8) & (vg.NA_ratio<0.2) & (vg.EU_ratio<0.2)& (vg.JP_ratio<0.2)]
Other_popular=Other_popular.sort_values(by="Other_ratio",ascending=False)
Other_popular
```

```python
data = Other_popular.head(5)
names = data['Name']
na = data['NA_ratio']
eu = data['EU_ratio']
jp = data['JP_ratio']
other = data['Other_ratio']

plt.figure(figsize=(7,5))

plt.bar(names, other, width=0.5, label="Other")
plt.bar(names, na, width=0.5, bottom=other, label="NA")
plt.bar(names, jp, width=0.5, bottom=other + na, label="JP")
plt.bar(names, eu, width=0.5, bottom=other + na + jp, label="EU")

plt.xlabel("Game Name")
plt.ylabel("Sales Ratio")
plt.title("Sales Performance of Other RegionsTop 5 Games Across NA,Europe and japan")
plt.xticks(rotation=90)
plt.legend()

plt.show()
```
